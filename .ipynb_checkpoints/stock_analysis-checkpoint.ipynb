```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("INFY.csv")
df.head()
```

```python
# Show data types and non-null counts
df.info()

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Remove rows with any missing values
df.dropna(inplace=True)

# Confirm it's cleaned
print("After cleaning, number of rows:", df.shape[0])
df.isnull().sum()
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Show the first few rows to confirm
df.head()
```

```python
# Calculate 20-day Moving Average of the Close price
df['MA20'] = df['Close'].rolling(window=20).mean()

# Show the last few rows to check the result
df.tail()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Close Price', color='blue')
plt.plot(df['MA20'], label='20-Day MA', color='orange')
plt.title('INFY Stock Price with 20-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
###  Project Summary: INFY Stock Market Analysis

- Loaded historical stock market data from a Kaggle dataset
- Cleaned missing values and converted the 'Date' column to datetime format
- Used NumPy and Pandas to calculate a 20-day Moving Average on closing prices
- Visualized stock price and trend using Matplotlib
- This project demonstrates basic data analysis and time-series plotting skills
```
